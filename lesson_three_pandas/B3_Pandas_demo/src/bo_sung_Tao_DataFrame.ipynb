### Tạo từ Series

```python
name_ser = pd.Series(['Bảo Ngọc','Thanh Bình','Ngọc Thủy','Đức Duy','Toàn Trung'],index=['a','b','c','d','e'])
salary_ser = pd.Series([2000,1200,1500,1700,1500],index=['a','b','c','d','f'])
department_ser = pd.Series(['KT','KT','IT','IT','KT'],index=['a','b','c','d','g'])
dic = {'ten':name_ser, 'luong':salary_ser, 'phong':department_ser}
employees_df = pd.DataFrame(dic)
employees_df
```
