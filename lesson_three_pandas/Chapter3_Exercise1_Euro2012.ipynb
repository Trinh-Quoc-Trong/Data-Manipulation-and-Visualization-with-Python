# Chapter 3 - Exercise 1: Lọc và sắp xếp dữ liệu UEFA_European_Championship/Euro 2012
Bộ dữ liệu này liên quan đến giải vô địch bóng đá châu Âu Euro 2012, bao gồm thông tin về các trận đấu, đội bóng, cầu thủ, kết quả, và nhiều yếu tố thống kê khác liên quan đến giải đấu Euro 2012.

#### Lấy dữ liệu từ đường dẫn
[Euro 12](https://raw.githubusercontent.com/jokecamp/FootballData/master/UEFA_European_Championship/Euro%202012/Euro%202012%20stats%20TEAM.csv) hoặc tập tin "euro2012.csv" cho sẵn trong thư mục data.

Dữ liệu bao gồm 16 quốc gia, với các cột như sau:
- **Team**: tên đội bóng
- **Goals**: Số bàn thắng đã ghi 
- **Shots on target**: Số lần sút mà thực sự bị thủ môn đối phương cản phá
- **Shots off target**: Số lần sút không trúng khung thành đối phương
- **Shooting Accuracy**: Tỷ lệ phần trăm giữa số lần sút trúng mục tiêu và tổng số lần sút. Đây là một chỉ số quan trọng để đánh giá độ chính xác của việc sút của đội bóng.
- **% Goals-to-shots**: Tỷ lệ phần trăm giữa số bàn thắng và tổng số lần sút. 
- **Total shots (inc. Blocked)**: Tổng số lần sút của đội bóng, bao gồm cả những lần sút bị chặn.
- **Hit Woodwork**: Số lần bóng của đội bóng đập vào khung gỗ mà không ghi bàn.
- **Penalty goals**: Số bàn thắng được ghi từ quả penalty.
- **Penalties not scored**: Số lần đá penalty nhưng không ghi bàn.
