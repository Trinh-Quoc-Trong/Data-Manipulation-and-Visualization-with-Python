```python
import pandas as pd  
from wordcloud import WordCloud, STOPWORDS  
import matplotlib.pyplot as plt  
from PIL import Image  
import numpy as np  
```

```python
file_path = 'ArticlesApril2017.csv'  
data = pd.read_csv(file_path)  
data
```

```python

data.info()
# data.shape
```

```python

missing_data = data.isnull().sum()  

print(missing_data)  
```

```python


headlines = data['headline'].dropna().astype(str) 
combined_text = ' '.join(headlines).lower()  
```

```python


stop_words = set(STOPWORDS)  
additional_stopwords = {'one', 'br', 'po', 'th', 'sayi', 'fo', 'unknown'}  
stop_words.update(additional_stopwords)  
```

Nhận xét:

Đã loại bỏ các từ không quan trọng được chỉ định trong từng tiêu đề.
Sử dụng một mảng stopwords để loại trừ những từ này khỏi văn bản ban đầu.
Kết quả là danh sách tiêu đề đã được chuẩn hóa.

## 2.	Tạo biểu đồ Wordcloud có số lượng từ tối đa là 50 và loại bỏ các từ trùng nhau như kết quả gọi ý sau : (0.5 điểm)

```python
wordcloud = WordCloud(  
        stopwords=stop_words,  
        max_words=50,  
        background_color="white",  
        colormap="viridis"  
    ).generate(combined_text)  
```

```python
plt.figure(figsize=(10, 7))  
plt.imshow(wordcloud, interpolation="bilinear")  
plt.axis("off")  
plt.title("Wordcloud từ dữ liệu headline", fontsize=16)  
plt.show()  
```

```python
bird_mask = np.array(Image.open("bird.jpg"))  

wordcloud_with_mask = WordCloud(  
    stopwords=stop_words,  
    mask=bird_mask,  
    contour_color='black',  
    contour_width=1,  
    background_color="white",  
).generate(combined_text)  

```

```python

plt.figure(figsize=(10, 7))  
plt.imshow(wordcloud_with_mask, interpolation="bilinear")  
plt.axis("off")  
plt.title("hình ảnh bird.jpg", fontsize=16)  
plt.show()  
```
